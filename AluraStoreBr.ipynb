# 2. Vendas por Categoria


```python
produtos = [
    ('Assistente virtual', 'eletronicos', 8),
    ('Mesa de jantar', 'moveis', 4),
    ('Jogo de tabuleiro', 'brinquedos', 1),
    ('Micro-ondas', 'eletrodomesticos', 1),
    ('Cadeira de escritório', 'moveis', 1)
]
quantidade_por_categoria = {}

for produto, categoria, quantidade in produtos:
    if categoria in quantidade_por_categoria:
        quantidade_por_categoria[categoria] += quantidade
    else:
         quantidade_por_categoria[categoria] = quantidade
print("Quantidade de produtos vendidos por categoria")
for categoria, quantidade in quantidade_por_categoria.items():
    print(f"{categoria}: {quantidade}")
```

# 4. Produtos Mais e Menos Vendidos

```python
dados = [
    ('Assistente virtual', 'Loja 1', 8),
    ('Mesa de jantar', 'Loja 2', 4),
    ('Jogo de tabuleiro', 'Loja 1', 1),
    ('Micro-ondas', 'Loja 3', 1),
    ('Cadeira de escritório', 'Loja 2', 1)
]
vendas_por_loja = {}
for produto, loja, quantidade in dados:
    if loja not in vendas_por_loja:
        vendas_por_loja[loja] = {}
    vendas_por_loja[loja][produto] = vendas_por_loja[loja].get(produto, 0) + quantidade
for loja, produtos in vendas_por_loja.items():
    produtos_ordenados = sorted(produtos.items(), key=lambda x: x[1], reverse=True)
    mais_vendido = produtos_ordenados[0]
    menos_vendido = produtos_ordenados[-1]

    print(f"Loja: {loja}")
    print(f"Mais vendido: {mais_vendido[0]} ({mais_vendido[1]} unidades)")
    print(f"Menos vendido: {menos_vendido[0]} ({menos_vendido[1]} unidades)")
    print("-" * 30)
```

```python
import matplotlib.pyplot as plt

# Lista de dados: (produto, quantidade)
dados = [
    ("Assistente virtual", 8),
    ("Mesa de jantar", 4),
    ("Jogo de tabuleiro", 1),
    ("Micro-ondas", 1),
    ("Cadeira de escritório", 1)
]

# Separando os nomes e as quantidades
produtos = [item[0] for item in dados]
quantidades = [item[1] for item in dados]

# Criando o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(quantidades, labels=produtos, autopct='%1.1f%%', startangle=90, colors=plt.cm.Pastel1.colors)
plt.title("Distribuição de Produtos Vendidos")
plt.axis('equal')  # Deixa o círculo redondinho

# Exibindo o gráfico
plt.show()
```
